## Matrix Operationen

Addition und Subtraktion funktionieren für Matrizen genau so wie wir das erwarten würden.
Für andere Operationen jedoch müssen wir, wenn wir nicht mit den einzelnen Elementen sondern der ***ganzen Matrix rechnen wollen, spezielle Operationen*** verwenden:

### Multiplication of Two Matrices
To multiply two matrices, we use `dot()` method.

````python
A = np.array([[3, 6, 7], [5, -3, 0]])
B = np.array([[1, 1], [2, 1], [3, -3]])
C = A.dot(B)
print(C)
````

Zwei mögliche Schreibweisen:

`A.dot(B)` : Methode .dot anwenden auf array A, mit B

`np.dot(A,B)` : np.dot Funktion (dot-product) ausführen mit arrays A und B

````python
A = np.array([[3, 6, 7], [5, -3, 0]])
B = np.array([[1, 1], [2, 1], [3, -3]])
D = np.dot(A,B)
print(D)
````

Achtung: Matrix-Multiplikation ist nicht kummulativ, d.h. ein vertauschen der Matrizen ergibt nicht mehr das gleiche Resultat

A.B =/= B.A

````python
A = np.array([[3, 6, 7], [5, -3, 0],[1, 2, 3]])
B = np.array([[1, 1, 6], [2, 1, -4],[3, -3, 7]])

C1 = A.dot(B)
print("A.B = ")
print(C1)
print("\n")

print("B.A = ")
C2 = B.dot(A)
print(C2)
````

"dot-product" ist ein anderer Name für das Skalarprodukt von Vektoren.
Und das Skalarprodukt ist seinerseits nichts anderes als eine Matrix-Multiplikation. Daher der Name dieser Operation hier.

**Simple Practical Application for Matrix Multiplication**

In the following practical example, we come to talk about the sweet things of life.

Let's assume there are four people, and we call them Lucas, Mia, Leon and Hannah. Each of them has bought chocolates out of a choice of three. The brand are A, B and C, not very marketable, we have to admit.



*   Lucas bought 100 g of brand A, 175 g of brand B and 210 of C.
*   Mia choose 90 g of A, 160 g of B and 150 g of C.
*   Leon bought 200 g of A, 50 of B and 100 g of C.
*   Hannah apparently didn't like brand B, because she hadn't bought any of those. But she she seems to be a real fan of brand C, because she bought 310 g of them. Furthermore she bought 120 g of A.

So, what's the price in CHF of these chocolates:

* A costs 2.98 per 100 g,
* B costs 3.90 and
* C only 1.99 CHF.

If we have to calculate how much each of them had to pay, we can use Python, NumPy and Matrix multiplication:

````python
# Informationen in Matrix & Vektor einlesen:
Person_Bought = np.array([[100, 175, 210],
                            [90, 160, 150],
                            [200, 50, 100],
                            [120, 0, 310]])
Price_per_100_g = np.array([2.98, 3.90, 1.99])
Price_per_1_g = Price_per_100_g / 100

# Kosten pro Person = Matrix.Vektor:
Price_in_CHF = np.dot(Person_Bought,Price_per_1_g)

# Resultat ausgeben:
print('Kosten für \n Lucas, Mia, Leon und Hanna in CHF: \n',Price_in_CHF)
````

### Find Multiplicative Inverse of a Given Matrix
Python provides a very easy method to calculate the inverse of a matrix. The function `numpy.linalg.inv()` which is available in the python NumPy module is used to compute the inverse of a matrix.

````python
# Taking a 3 * 3 matrix
A = np.array([[6, 1, 1],
              [4, -2, 5],
              [2, 8, 7]])

# Calculating the inverse of the matrix
AI = np.linalg.inv(A)
print(AI)
````

Inverse ausprobieren: Matrix-Multiplikation = Einheit???

````python
print(np.dot(AI,A))
````

````python
print(np.dot(A,AI))
````

# Teil 2: Anwendung Gleichungssysteme

Die Inverse einer Matrix brauchen wir zum Beispiel, um ganz effizient riesige lineare Gleichungssysteme zu lösen.

**einfaches Beispiel**

2x2 Gleichungssystem:



```
1x + 2y = 4
0x + 3y = 7
```



in Matrix Form:

image.png

$\mathbf{A}\cdot \vec{x} = \vec{c}$

Ganz naiv können wir diese Gleichung nun lösen, als ob das nur eine einzige Gleichung wäre, und nicht zwei gekoppelte Gleichungen mit zwei Variablen.

Auf beiden Seiten mit der Inversen von $\mathbf{A}$ multiplizieren:

$\vec{x} = \mathbf{A}^{-1} \cdot \vec{c}$

Dh die Lösung des Gleichungssystems finden wir direkt mit:

````python
# Matrix A und Vektor c eingeben:
A = np.array([[1,2],
             [0,3]])
c = np.array([[4],
              [7]])

# Lösung berechnen und ausgeben:
x = np.linalg.inv(A).dot(c)
print(x)
````

Diese Methode funktioniert, egal wie gross das zu lösende lineare Gleichungssystem ist, solange eine korrekte Lösung existiert.

# Teil 4: Anwendung Stochastische Prozesse - Beispiel Diffusion

image.png

Die Verteilung der Teilchen auf die beiden Hälften ändert sich jeweils nach Ablauf einer
festen Zeiteinheit:
* 10 % der sich in A befindlichen Teilchen gelangen nach B (die restlichen 90 %
bleiben in A) und
* 20 % der sich in B befindlichen Teilchen gelangen nach A (die restlichen 80 %
bleiben in B).

Am Anfang sind
* 3000 Teilchen in A und
* 9000 Teilchen in B.

Es soll untersucht werden, wie sich die Verteilung der 12000 Teilchen auf die beiden Hälften entwickelt.


**Zustandsvektor**

Vektor beschreibt Anzahl Teilchen in den beiden Teilen, zu einem bestimmten Zeitpunkt

**stochastische Matrix**

beschreibt Austauschprozess der Teilchen zwischen A und B

*   quadratisch
*   keine negative Einträge
*   Summe in Spalten = 1

Darstellung in Matrix:
* Spalte = Ursprung (von)
* Zeile = Ziel (nach)

image.png

Zustand nach einem Zeitschritt = Matrixmultiplikation

````python
# Start-Zustand
Z0 = np.array([[3000],
               [9000]])

# stochastische (Übergangs-)Matrix
U = np.array([[0.9, 0.2],
              [0.1, 0.8]])

# Zustand nach einem Zeitschritt:
Z1 = np.dot(U,Z0)
print(Z1)
````

d.h. neu sind 4500 Teilchein in A, und 7500 in B.

**WIE KANN MAN DEN ZUSTAND NACH 2, 3, ODER n ZEITSCHRITTEN FINDEN?**

````python
# Zustand nach 2 Zeitschritten:
Z2 = np.dot(U,Z1)
print(Z2)

# Zustand nach 3 Zeitschritten:
Z3 = np.dot(U,Z2)
print(Z3)
````

Auch für Matrizen kann man eine Potenz-Operation definieren.
so lange die Matrix quadratisch ist.

zum Beispiel: $M^3 = M\cdot M \cdot M$

in python:

`numpy.linalg.matrix_power(M,n)`


````python
Z10 = np.linalg.matrix_power(U,10).dot(Z0)
print(Z10)
````
